## Heap

```python
import graphviz
import IPython
```

```python
from math import inf, nan
def new_heap():
    return [nan]

def heap_push(heap, x):
    heap.append(x)
    cur = len(heap) - 1
    while x < heap[cur // 2]:
        heap[cur], heap[cur // 2] = heap[cur // 2], heap[cur]
        cur = cur // 2

def heap_top(heap):
    if len(heap) > 1:
        return heap[1]
    return None

def heapify(heap, cur):
    while 2 * cur < len(heap):
        target = 2 * cur
        if target + 1 < len(heap) and heap[target+1] < heap[target]:
            target += 1
        if heap[target] < heap[cur]:
            heap[target], heap[cur] = heap[cur], heap[target]
            cur = target
        else:
            return
    return

def heap_pop(heap):
    if len(heap) < 2:
        return
    heap[1], heap[-1] = heap[-1], heap[1]
    heap.pop()
    heapify(heap, 1)

def heap_is_empty(heap):
    return len(heap) <= 1

def make_heap(array):
    for i in range(len(array) // 2, 0, -1):
        heapify(array, i)
```

```python
h = new_heap()
```

```python
for i in range(10, 0, -1):
    heap_push(h, i)

print(h)
```

```python
while not heap_is_empty(h):
    print(heap_top(h))
    heap_pop(h)
```

```python
def heap_to_tree(heap):
    g = graphviz.Digraph()
    if not heap_is_empty(heap):
        q = [1]
        g.node(str(heap[1]))
        while len(q) > 0:
            cur = q.pop(0)
            if cur * 2 < len(heap):
                q.append(cur * 2)
                g.edge(str(heap[cur]), str(heap[cur * 2]))
            if cur * 2 + 1 < len(heap):
                q.append(cur * 2 + 1)
                g.edge(str(heap[cur]), str(heap[cur * 2 + 1]))
    return g
```

```python
h = new_heap()
for i in range(5, 0, -1):
    heap_push(h, i)
    IPython.display.display(heap_to_tree(h))
    print(h)
```

```python
def heap_push_display(heap, x):
    IPython.display.display(heap_to_tree(heap))
    heap.append(x)
    cur = len(heap) - 1
    while x < heap[cur // 2]:
        IPython.display.display(heap_to_tree(heap))
        print(heap)
        heap[cur], heap[cur // 2] = heap[cur // 2], heap[cur]
        cur = cur // 2
    IPython.display.display(heap_to_tree(heap))
    print(heap)

h = new_heap()
for i in range(7, 1, -1):
    heap_push(h, i)

heap_to_tree(h)
print(h)

heap_push_display(h, 1)
```

```python
array = [nan] + [i for i in range(10, 0, -1)]
IPython.display.display(heap_to_tree(array))
print(array)
make_heap(array)
IPython.display.display(heap_to_tree(array))
print(array)
```
